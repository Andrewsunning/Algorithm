# 一、排序
### 1.快速排序
```
给定你一个长度为n的整数数列。

请你使用快速排序对这个数列按照从小到大进行排序。

并将排好序的数列按顺序输出。

输入格式
输入共两行，第一行包含整数 n。

第二行包含 n 个整数（所有整数均在1~109范围内），表示整个数列。

输出格式
输出共一行，包含 n 个整数，表示排好序的数列。

数据范围
1≤n≤100000
输入样例：
5
3 1 2 4 5
输出样例：
1 2 3 4 5
```

````python
"""
基本思想：递归，分而治之，双指针
    1. 确定分界点:mid = alist[(l+r)//2]
    *2. 调整区间，使<mid的元素在alist左面，使>mid的元素在alist右面
    3. 递归处理左右半边

空间复杂度：O(logn - n)
最优时间复杂度：O(n*log n)
最坏时间复杂度：O(n**2)
平均时间复杂度：O(n*log n)
稳定性：不稳定（为什么？？？）

！注意：整个过程操作的是同一个列表，而不是列表的copy
"""

def quickSort(alist, left, right):
    if left>=right:
        return
    
    i = left-1
    j = right+1
    mid = alist[(left+right)//2]    # 注意，这种写法不能取mid = alist[left]作为mid，否则会陷入死循环
    
    while i<j:
        while 1:
            i+=1
            if alist[i]>=mid:
                break
        while 1:
            j-=1
            if alist[j]<=mid:
                break
        if i<j:
            alist[i], alist[j] = alist[j], alist[i]
    quickSort(alist,left, j)    # 参数必须是j，i的话会产生边界问题
    quickSort(alist,j+1,right)

if __name__=="__main__":
    n = int(input().strip())
    alist = list(map(int, input().split()))
    
    quickSort(alist, 0, n-1)
    
    [print(i, end=" ") for i in alist]
````

### 2. 归并排序
```
给定你一个长度为n的整数数列。

请你使用归并排序对这个数列按照从小到大进行排序。

并将排好序的数列按顺序输出。

输入格式
输入共两行，第一行包含整数 n。

第二行包含 n 个整数（所有整数均在1~109范围内），表示整个数列。

输出格式
输出共一行，包含 n 个整数，表示排好序的数列。

数据范围
1≤n≤100000
输入样例：
5
3 1 2 4 5
输出样例：
1 2 3 4 5
```

````python
"""
基本思想:
   1. 确定分界点索引,mid=(l+r)//2
   2. 递归排序[left,mid]和[mid+1,right]两部分
   3. *归并：按序合并[left,mid]和[mid+1,right]两部分到tmp数组
"""
def mergeSort(alist, l, r):
    if l>=r: return
    
    mid = (l+r)//2
    mergeSort(alist, l, mid)
    mergeSort(alist, mid+1, r)
    
    # 合并：遍历比较alist数组中[l,mid]和[mid+1,r]两部分的元素，小的元素存到tmp中
    k = 0    # 存tmp数组中元素的个数
    i = l   # 左半部分的指针
    j = mid+1    # 右半部分的指针
    while i<=mid and j<=r:
        if alist[i]<alist[j]:
            tmp[k] = alist[i]
            k+=1
            i+=1
        else:
            tmp[k] = alist[j]
            k+=1
            j+=1
    # 将合并的剩余部分的数组元素存到tmp的后面
    while i<=mid:
        tmp[k] = alist[i]
        k+=1
        i+=1
    while j<=r:
        tmp[k] = alist[j]
        k+=1
        j+=1
    
    # 将排序好的tmp数组存到alist对应部分中
    i = l    # alist的指针，从left开始
    j = 0    # tmp的指针，从0开始
    while i<=r:
        alist[i] = tmp[j]
        i+=1
        j+=1
    
if __name__=="__main__":
    n = int(input().strip())
    nums = list(map(int, input().split()))
    
    tmp = [0]*n    # 一维数组可以这么生成，二维数组不行
    mergeSort(nums, 0, n-1)
    
    [print(item, end=" ") for item in nums]
````

### 2.2 逆序对的数量
```
给定一个长度为n的整数数列，请你计算数列中的逆序对的数量。

逆序对的定义如下：对于数列的第 i 个和第 j 个元素，如果满足 i < j 且 a[i] > a[j]，则其为一个逆序对；否则不是。

输入格式
第一行包含整数n，表示数列的长度。

第二行包含 n 个整数，表示整个数列。

输出格式
输出一个整数，表示逆序对的个数。

数据范围
1≤n≤100000
输入样例：
6
2 3 4 5 6 1
输出样例：
5
```

````python
"""
基本思想：
    借助归并排序模板，在合并过程中：
    1. 如果nums[i]<=nums[j]，即nums[i]与nums[j]没有构成逆序对，则正常进行合并操作
    2. 如果左半部分的nums[i]>右半部分的nums[j],那么i后面的元素都大于nums[j],即逆序对个数为(mid-i+1)
"""
def mergeSort(l, r):
    if l>=r: return 0    # 如果子数组里有之多一个元素，那么逆序对个数为0
    mid = (l+r)//2
    res = mergeSort(l, mid) + mergeSort(mid+1, r)
    k = 0
    i = l
    j = mid+1
    while i<=mid and j<=r:
        if nums[i]<=nums[j]:    # nums[i]与nums[j]没有构成逆序对
            tmp[k] = nums[i]
            k+=1
            i+=1
        else:    # 如果左半部分的nums[i]>右半部分的nums[j],那么i后面的元素都大于nums[j],即逆序对个数为(mid-i+1)
            res += mid-i+1
            tmp[k] = nums[j]
            k+=1
            j+=1
    while i<= mid:
        tmp[k] = nums[i]
        k+=1
        i+=1
    while j<=r:
        tmp[k] = nums[j]
        k+=1
        j+=1
    i = l
    j = 0
    while i<=r:
        nums[i] = tmp[j]
        j+=1
        i+=1
    return res
    

if __name__=="__main__":
    n = int(input().strip())
    nums = list(map(int, input().split()))
    tmp = [0]*n
    
    res = mergeSort(0,n-1)
    print(res)
````

### 3.数的范围
```
给定一个按照升序排列的长度为n的整数数组，以及 q 个查询。

对于每个查询，返回一个元素k的起始位置和终止位置（位置从0开始计数）。

如果数组中不存在该元素，则返回“-1 -1”。

输入格式
第一行包含整数n和q，表示数组长度和询问个数。

第二行包含n个整数（均在1~10000范围内），表示完整数组。

接下来q行，每行包含一个整数k，表示一个询问元素。

输出格式
共q行，每行包含两个整数，表示所求元素的起始位置和终止位置。

如果数组中不存在该元素，则返回“-1 -1”。

数据范围

1≤q≤10000
1≤k≤10000
输入样例：
6 3
1 2 2 3 3 4
3
4
5
输出样例：
3 4
5 5
-1 -1
```

````python
"""
二分查找的本质：是找到边界点，使一侧满足check条件，而另一侧不满足
"""
def binarySearch_1(l, r):
    global k
    while l<r:
        mid = (l+r)//2
        if nums[mid]>=k: r = mid
        else: l = mid+1
    return l

def binarySearch_2(l, r):
    global k

    while l<r:
        mid = (l+r+1)//2
        if nums[mid]<=k: l = mid
        else: r = mid-1
    return l

if __name__=="__main__":
    n, q = map(int,input().split())
    nums = list(map(int, input().split()))
    
    while q:
        k = int(input().strip())
        l = binarySearch_1(0, n-1)
        if nums[l]!=k:
            print("-1 -1")
        else:
            print(l, end=" ")
            r = binarySearch_2(0, n-1)
            print(r)
        q-=1
````

#### 旋转数组
```
把一个数组最开始的若干个元素搬到数组的末尾，我们称之为数组的旋转。
输入一个非递减排序的数组的一个旋转，输出旋转数组的最小元素。
例如数组{3,4,5,1,2}为{1,2,3,4,5}的一个旋转，该数组的最小值为1。
NOTE：给出的所有元素都大于0，若数组大小为0，请返回0。
```

````python

# -*- coding:utf-8 -*-
class Solution:
    def minNumberInRotateArray(self, rotateArray):
        # write code here
        if not rotateArray:
            return 0
        start = 0
        end = len(rotateArray) - 1
        while start<=end:
            mid = (start+end)//2
            if rotateArray[mid]<rotateArray[mid-1]:    # 若mid个元素小于mid-1个元素，则mid个元素为最小值
                return rotateArray[mid]
            elif rotateArray[mid]<=rotateArray[end]:    # 若第mid个元素小于等于第end个元素，则最小元素在mid的左侧，eg[1,1,1,0,1,1]
                end = mid - 1
            else:    # 若第mid个元素大于第end个元素，则最小元素在mid的右侧,eg[6,1,2,3,4,5]
                start = mid + 1
        return 0
````

### 7.差分
```
输入一个长度为n的整数序列。

接下来输入m个操作，每个操作包含三个整数l, r, c，表示将序列中[l, r]之间的每个数加上c。

请你输出进行完所有操作后的序列。

输入格式
第一行包含两个整数n和m。

第二行包含n个整数，表示整数序列。

接下来m行，每行包含三个整数l，r，c，表示一个操作。

输出格式
共一行，包含n个整数，表示最终序列。

数据范围
1≤n,m≤100000,
1≤l≤r≤n,
−1000≤c≤1000,
−1000≤整数序列中元素的值≤1000
输入样例：
6 3
1 2 2 1 2 1
1 3 1
3 5 1
1 6 1
输出样例：
3 4 5 3 4 2
```

````python
# @pysnooper.snoop()
def insert(l, r, c):
    """相当于对前缀和数组的[l,r]范围元素加上c"""
    b[l]+=c
    b[r+1]-=c

if __name__=="__main__":
    n,m = map(int, input().split())
    
    a = [0]*(n+1)    # 初始化前缀和数组
    b = [0]*(n+2)    # 初始化差分数组，！！！注意：数组的首位都要填充一个元素，不然当r=len(a)时，insert()函数中r+1会越界
    in_li = list(map(int, input().split()))
    for i in range(1, n+1):    # 1. 读入前缀和数组
        a[i] = in_li[i-1]
    
    for i in range(1,n+1):    # 2. 构造差分数组
        insert(i, i, a[i])
        
    while m:    # 3. 对差分数组进行m步操作
        l, r, c = map(int, input().split())
        insert(l, r, c)
        m-=1
        
    s = [0]*(n+1)    # 4. 计算操作后的b数组的前缀后
    for i in range(1, n+1):
        s[i] = s[i-1]+b[i]
    [print(s[i], end=' ') for i in range(1,n+1)]
````

给定一个长度为n的数列，请你求出数列中每个数的二进制表示中1的个数。

输入格式
第一行包含整数n。

第二行包含n个整数，表示整个数列。

````python
"""
位运算的常用操作：
1. 求n的二进制表示中从右至左的第k位数字
n >> k & 1

2. 返回n的二进制表示中从左至右的最后一位1的位置
lowbit(x) = x & -x
"""
def lowbit(x):
    """返回x的二进制表示中从左至右的最后一位1的位置"""
    return x & -x
    
if __name__=="__main__":
    n = int(input().strip())
    
    nums = list(map(int, input().split()))
    
    for i in range(n):
        res = 0
        x = nums[i]
        while x:
            x-=lowbit(x)
            res+=1
        print(res, end=" ")
````
